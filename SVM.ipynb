```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
iris.info()
```

```python
x=iris.iloc[0:100,0:2]
y=iris.iloc[0:100,-1]
print(x)
print(y)
```

```python
sc=StandardScaler()
xs=sc.fit_transform(x)
```

```python
from sklearn.svm import SVC
svc=SVC(C=1.0,kernel='linear',degree=3,gamma='auto')
model=svc.fit(xs,y)
```

```python
clr=['green' if c=='versicolor' else 'blue' for c in y]
plt.scatter(xs[:,0],xs[:,1], c=clr)
```

```python
w=svc.coef_
print(w)
```

```python
clr=['green' if c=='versicolor' else 'blue' for c in y]
plt.scatter(xs[:,0],xs[:,1], c=clr)

#hyperplane
w=svc.coef_[0]
a=-w[0] / w[1]
xx=np.linspace(-2.5,2.5)
yy=a*xx - (svc.intercept_[0]) / w[1]
plt.plot(xx,yy)
plt.axis("off"),plt.show()
```
