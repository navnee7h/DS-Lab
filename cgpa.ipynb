```python
import pandas as pd
import numpy as np
student={'id':[1,2,3,4,5,6],
        'name':['A','B','C','D','E','F'],
        'age':[21,20,22,23,20,np.nan],
        'degree':['mca','msc','mca','msc','mca','msc'],
        'cgpa':[7.5,8,9,8,9,np.nan]}
df=pd.DataFrame(student)
print(df)
print("a)insert a new column ‘class’ into the dataframe and populate values")
clas = ['first','second','first','first','second','second']
df['class']=clas
print(df)
print("b)Find out total no.of columns in the dataframe")
tot_col=len(df.axes[1])
print(tot_col)
print("c)select top 3 rows ,last 4 rows")
print(df.head(3))
print(df.tail(4))
print("g)fill null values in the age column with mean value")
mean=df['age'].mean()
df['age'].fillna(value=mean, inplace=True)
print(df)
print("d)delete all rows which have null values")
df = df.dropna(how='any',axis=0) 
print(df)
print("e)drop ‘id’ column")
df=df.drop(['id'],axis=1)
print(df)
print("f)select name,cgpa from 2 to 5 records")
df=df.iloc[1:5,[0,3]]
print(df)
```

```python
#22
import pandas as pd
print("(a)Load iris dataset into a dataframe")
df=pd.read_csv('iris.csv')
df=pd.DataFrame(df)
print(df)
print("(b)Take its descriptive statistics")
mean=df.mean()
print("Mean is:",mean)
mode=df.mode()
print("Mode is:")
print(mode)
median=df.median()
print("Median is:")
print(median)
print("(c)delete all rows which have null values")
df=df.dropna(how="any",axis=0)
print(df)
print("(d)select the Species classes,Count the no.of datas in each class")
dd=df["species"].value_counts()
print(dd)
print("(e)Map the species classes into 1,2 and 3")
print("(f)Print the standard deviation and mean of petal length")
dm=df['petal_length'].mean()
print("Mean is:",dm)
ds=df['petal_length'].std()
print("Std is:",ds)
print("(g)select all columns except species")
df1 = df.loc[:, df.columns != "species"]
print(df1)

```
