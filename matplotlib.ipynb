Programs using matplotlib /  seaborn for data visualisation
a)Write a program to draw univariate visualization plots(line plot,histogram,boxplot,barchart,piechart) with matplotlib for iris dataset
b)Write a program to draw multivariate visualization plots(scatter plots, scatter multiple,scatter matrix,bubble plot) with matplotlib for iris dataset
c)Write a program to draw univariate and multivariate visualization plots with seaborn(line plot, pairplot,jointplot,heatmap) for iris dataset


```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('iris.csv')
print(df)
x=df["sepal_length"]
plt.hist(x,bins=20,color='green')
plt.title('Sepal Length')
plt.xlabel('sepal_length')
plt.ylabel('count')
```

```python
df.plot(kind = 'line')
df.plot(kind = 'line',subplots=True)
df["sepal_length"].plot(kind = 'line')
```

```python
df.boxplot()
```

```python
df.boxplot(["sepal_length"])
```

```python
df["sepal_width"].plot(kind = 'box')
```

```python
df["sepal_length"].plot.bar(bin=20,color="red")
```

```python
plt.pie(df['species'].value_counts(),labels=['setosa','versicolor','virginica'],shadow=True)
```

```python
df.plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width')
df.plot.scatter(x='petal_length', y = 'petal_width')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('iris.csv')
ax1 = df.plot(kind='scatter', x='petal_length', y = 'petal_width', label='petal_length Vs petal_width', color='r') 
ax2 = df.plot(kind='scatter', x='petal_length', y = 'sepal_length', label='petal_length Vs sepal_length',color='g',ax=ax1)
ax3 = df.plot(kind='scatter',  x='petal_length', y = 'sepal_width',label='petal_length Vs sepal_width', color='b',ax=ax2)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df, alpha=0.5, diagonal='kde')
pd.plotting.scatter_matrix(df)
```

```python
plt.scatter('petal_length', 'petal_width',  s='sepal_length',  alpha=0.5,  data=df)
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
sns.lineplot(data=iris)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,height=4)
```

```python
sns.pairplot(data=iris,kind='scatter')
```

```python
sns.heatmap(iris.corr(),annot=True)
```
