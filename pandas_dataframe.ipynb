19.Write a pandas program to create a dataframe(6 records) which has the fields id,name,age,degree,cgpa
a)insert a new column ‘class’ into the dataframe and populate values
b)Find out total no.of columns in the dataframe
c)select top 3 rows ,last 4 rows
d)delete all rows which have null values
e)drop ‘id’ column
f)select name,cgpa from 2 to 5 records
g)fill null values in the age column with mean value

```python
import pandas as pd
import numpy as np
data={'ID':[1,2,3,4,5,6],
      'Name':['Ram','Shyam','Sam','Tom','John','Jacob'],
      'Age':[21,22,19,np.nan,24,23],
      'Degree':['CS','Physics','Maths','BCom','MCA','BTech'],
      'CGPA':[8,9,10,6,7,9]}
df=pd.DataFrame(data)
clas=['First Class','First Class','Distinction','Second Class','Second Class','First Class']
df=df.assign(Class=clas)
print(df)
print('\n Total number of Columns in the Data frame :',len(df.axes[1]))
print('\n Top 3 rows :\n',df.head(3))
print('\n Last 4 rows :\n',df.tail(3))
df1=df.dropna()
print('\n After deleting rows with null values :\n',df1)
df.pop('ID')
print('\n After dropping column-ID :\n',df)
print('\n Name and Cgpa from 2 to 5 records :\n',df.iloc[[2,3,4,5],[0,3]])
df['Age'].fillna(value=df['Age'].mean(),inplace=True)
print('\n After filling null values in the age column with mean value :\n',df)
```

20.a)Load iris dataset into a dataframe
b)Take its descriptive statistics
c)delete all rows which have null values
d)select the Species classes,Count the no.of datas in each class
e)Map the species classes into 1,2 and 3
f)Print the standard deviation and mean of petal length
g)select all columns except species


```python
import pandas as pd
df=pd.read_csv('iris.csv')
print(df)
print('\n Descriptive Statistics :\n\n',df.describe())
print('\n After deleting rows which have null values :\n')
df1=df.dropna()
print(df1)
print('\n Species Classes & number of data in each class :\n')
print(df.groupby(['species']).count())
print('\nStandard deviation of Petal length :')
print(df['petal_length'].std())
print('\nMean of Petal length :')
print(df['petal_length'].mean())
```
