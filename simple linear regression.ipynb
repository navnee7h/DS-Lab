```python
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
df=pd.read_csv("Salary_Data.csv")
```

```python
#check the correleation between dependent and independent variable
import seaborn as sb
sb.heatmap(df.corr(),annot=True)
```

```python
#seperatinng the dependent and independent variable
x=df.iloc[:, :-1].values
print(x)
y=df.iloc[:,1].values
print(y)
```

```python
#spliting the dataset to training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
#fitting regression model into training and testing data set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#prediicting the result
y_pred=regressor.predict(x_test)
x_pred=regressor.predict(x_train)
```

```python
#ploting the scatter plot
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.plot(x_train,x_pred,color="Black")
plt.xlabel("years of experience")
plt.ylabel("salary")
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color="black")
plt.title("Scatter plot of testing")
plt.xlabel("years of experience")
plt.ylabel("salary")
```

```python
print(regressor.score(x_test,y_test))
```

```python
print(regressor.score(x_train,y_train))
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
